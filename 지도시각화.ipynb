```python
import pandas as pd 

happy = pd.read_csv('all_Happy.csv')

happy.head(3)
```

```python

# 대한민국 시도별 위도, 경도 데이터를 DataFrame 형태로 생성
data = {
    "시도": ["서울특별시", "부산광역시", "대구광역시", "인천광역시", "광주광역시", 
              "대전광역시", "울산광역시", "세종특별자치시", "경기도", "강원도", 
              "충청북도", "충청남도", "전라북도", "전라남도", "경상북도", 
              "경상남도", "제주특별자치도"],
    "위도": [37.5665, 35.1796, 35.8714, 37.4563, 35.1595, 
            36.3504, 35.5384, 36.4800, 37.4138, 37.8228, 
            36.8000, 36.5184, 35.7175, 34.8679, 36.5760, 
            35.4606, 33.4996],
    "경도": [126.9780, 129.0756, 128.6014, 126.7052, 126.8526, 
            127.3845, 129.3114, 127.2890, 127.5183, 128.1555, 
            127.7000, 126.8000, 127.1530, 126.9910, 128.5056, 
            128.2132, 126.5312]
}

# DataFrame 생성
df = pd.DataFrame(data)
df.head()

```

```python
#두개 정보를 시도로 병합 함
df_merged = pd.merge(happy,df, on='시도')
df_merged.head()
```

```python
import folium

# 지도 객체 생성 (대한민국 중심으로 설정)
map_korea = folium.Map(location=[36.5, 127.5], zoom_start=7)

# 각 시도별로 원을 추가
for index, row in df_merged.iterrows():
    # 삶의 만족도를 기반으로 원의 크기 결정 (기본 크기에 삶의 만족도를 곱함)
    radius = 50000 * row['삶의 만족도'] # 원의 크기 조정을 위한 비례 계수

    # 원 추가
    folium.Circle(
        location=[row['위도'], row['경도']],
        radius=radius,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"{row['시도']}<br>삶의 만족도: {row['삶의 만족도']:.2f}"
    ).add_to(map_korea)

# 지도 표시
map_korea

```
